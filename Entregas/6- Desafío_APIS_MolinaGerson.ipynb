Para este desafío contemplamos una API publica del gobierno de Mexico referente a datos de condiciones atmosfericas.

El archivo está contemplado en formato json tal cómo lo indica el desafío a presentar. 

```python
import pandas as pd
import requests

url = 'https://api.datos.gob.mx/v1/condiciones-atmosfericas'

# Realizar la solicitud a la URL
response = requests.get(url)

# Verificar si la solicitud fue exitosa (código de respuesta 200)
if response.status_code == 200:
    # Cargar el contenido JSON
    data = response.json()

    # Extraer la lista de resultados
    results_list = data['results']

    # Convertir la lista de resultados en un DataFrame
    df = pd.DataFrame(results_list)

    # Mostrar el DataFrame
    print(df)
else:
    print(f"Error al realizar la solicitud. Código de respuesta: {response.status_code}")
```
