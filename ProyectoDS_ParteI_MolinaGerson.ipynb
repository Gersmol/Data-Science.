```python
#Se llenó los valores nulos con la sintaxis 'No especificado'.
df['Courier Status'].fillna('No especificado', inplace=True)
df['ship-city'].fillna('No especificado', inplace=True)
df['ship-state'].fillna('No especificado', inplace=True)
df['ship-postal-code'].fillna('No especificado', inplace=True)
df['ship-country'].fillna('No especificado', inplace=True)
df['promotion-ids'].fillna('No especificado', inplace=True)

#Se eliminaron estas filas con valores nulos ya que no se consideraron indispensables para el análisis.
df.dropna(subset=['currency', 'Amount'], inplace=True)

#Se consideró llenar los valores nulos con la moda.
mode_b2b = df['B2B'].mode()[0]
df['B2B'].fillna(mode_b2b, inplace=True)
mode_fulfilled_by = df['fulfilled-by'].mode()[0]
df['fulfilled-by'].fillna(mode_fulfilled_by, inplace=True)
```

ABSTRACT:

Este proyecto se enfoca en analizar un conjunto de datos que recopila información sobre las ventas de Amazon en el año 2022 en la India. El conjunto de datos incluye detalles como el número de orden, la fecha de la transacción, el estado del pedido, el canal de ventas, el nivel de servicio de envío, el estilo del producto, el SKU, la categoría y otros atributos relacionados con las ventas. En total, el conjunto de datos contiene 121.180 registros.
Al analizar los datos, se encontraron varios puntos de interés. En términos de ventas totales suma l78.592.678 INR. Las ventas se distribuyeron en todo el país, con ciudades como Bengaluur, Hyderabad y Mumbyu destacándose como destinos de envío más comunes.
El cumplimiento de los pedidos fue gestionado tanto por Amazon como por vendedores externos (Merchant)C con Amazin siendo el más prominente. Los productos vendidos abarcaron diversas categorías, desde kurtas hasta vestidos occidentales topsps.
Los datos revelaron la presencia de pedidos cancelados, aunque en menor medida en comparación con los pedidos enviados. Las promociones y las ventas B2B también desempeñaron un papel en algunas transacciones.
En resumen, este análisis proporciona una visión general de las ventas de Amazon en  en la India 2022, destacando tendencias clave en términos de ventas totales, ubicaciones de envío y canales de ventas utilizados. Los resultados pueden ser de interés para futuras estrategias comerciales y decisiones relacionadas con el comercio electrónico.

```python
plt.figure(figsize=(10, 6))
ax = sns.countplot(data=df, x='Category', palette='Set1')
plt.xticks(rotation=90)
plt.xlabel('Categoría de Producto')
plt.ylabel('Cantidad de Ventas')
plt.title('Ventas por Categoría de Producto')

for p in ax.patches:
    ax.annotate(format(p.get_height(), '.0f'), 
                (p.get_x() + p.get_width() / 2., p.get_height()), 
                ha='center', va='baseline', fontsize=10, color='black')
plt.show()
```

Las categorías SET y KURTA son claramente las categorías de mayor cantidad de ventas por lo que generan la mayor parte de los ingresos del negocio y una significativa demanda de los clientes. Sería fundamental mantener un enfoque constante en la gestión y promoción de estos productos. En el caso de las categorías Dupatta,Saree,Bottom y Blouse, podríamos considerar acciones estrátegicas de marketing, precios y calidad de los productos para aumentar su rotación.

```python
ventas_por_ciudad = df.groupby('ship-city')['Qty'].sum().reset_index()
ventas_por_ciudad = ventas_por_ciudad.sort_values(by='Qty', ascending=False)
ventas_por_ciudad_top_10 = ventas_por_ciudad.head(10)
ventas_por_ciudad_top_10
```

```python
plt.figure(figsize=(10, 6))
plt.bar(ventas_por_ciudad_top_10['ship-city'], ventas_por_ciudad_top_10['Qty'], color='skyblue')

for i, v in enumerate(ventas_por_ciudad_top_10['Qty']):
    plt.text(i, v + 50, str(v), ha='center', va='bottom')

plt.xlabel('Ciudad')
plt.ylabel('Cantidad de Ventas')
plt.title('Top 10 Ciudades por Cantidad de Ventas')
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()
```

En términos de venta, Bengaluru lidera el mercado, a pesar que esta ciudad es el lider Hyderabad, Mumbai, New Delhi y Chenai también contribuyen significativamente a las ventas. Esto demuestra que la demanda de los productos varía según su ubicación Geográfica lo que podría estar relacionado a estacionalidad, estilo de vida e incluso factores económicos. 

```python
sns.kdeplot(df['Amount'], shade=True, color='skyblue')
plt.xlabel('Monto')
plt.ylabel('Densidad')
plt.title('Densidad por Monto de Transacción')
plt.show()
```

Podemos inferír en que la Mayoría de transacciones se concentran apróximadamente entre 200 y 800 INR concentrando la mayor proporción de su ticket de ventas en 500 INR apróximadamente.

```python
productos_por_cantidad = df.groupby('SKU')['Qty'].sum()
productos_por_valor= df.groupby('SKU')['Amount'].sum()

plt.figure(figsize=(10, 6))
plt.scatter(productos_por_cantidad, productos_por_valor, alpha=0.5)
plt.title('Relación entre Cantidad Vendida y Valor de Ventas')
plt.xlabel('Cantidad Vendida')
plt.ylabel('Valor de Ventas')
plt.grid(True)
plt.show()
```

Podemos determinar en primera instancia una fuerte relación positiva entre la cantidad de ventas y el valor de las mismas, lo que podría significar que a medida que la cantidad aumenta, el valor de las ventas tienden a aumentar positivamente. Concluyendo qué, mientras mas unidades venda genero mas ingresos por esas ventas.
